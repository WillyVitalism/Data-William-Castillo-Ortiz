<h3 Style = "color: #8FBC8F"> 1.3 Revisión de encabezados: </h3>

<ul>
    <li>Nombre de columnas no deben contener espacios en blanco.</li>
    <li>Nombres de columnas no deben comenzar con números.</li>
    <li>No deben existir nombres repetidos.</li>
    <li>Recomendación, la primer palabra deber ser en minuscula.</li>
</ul>

```python
df_1er_autos.rename({"engine-size":"engineSize"}, axis = 1, inplace = True)
df_1er_autos.rename({"Cylinders":"cylinders"}, axis = 1, inplace = True)
df_1er_autos.rename({"City":"city"}, axis = 1, inplace = True)
df_1er_autos.rename({"Higway":"higway"}, axis = 1, inplace = True)
df_1er_autos.columns
```

```python
dfAutos.replace ("?", np.nan, inplace=True)
dfAutos.head (5)
```

<h3 Style = "color: #8FBC8F"> 2.2 Contar la cantidad de valores faltantes por variable (columnas): </h3>

```python
dfTemp = dfAutos.isnull()
dfTemp.head (5)
```

```python
for columna in dfTemp.columns.values.tolist ():
    print (columna)
    print (dfTemp[columna].value_counts())
    print ("")
```

<h3 Style = "color: #8FBC8F"> 2.3 Imputación simple de valores númericos: </h3>

```python
promNormalizadas = dfAutos ["normalized"].astype ("float").mean (axis = 0)
promNormalizadas
```

```python
dfAutos ["normalized"].replace (np.nan, 122, inplace = True)
dfAutos.head (5)
```

```python
prom = dfAutos ["bore"].astype ("float").mean (axis = 0)
dfAutos ["bore"].replace (np.nan, prom, inplace = True)
prom = dfAutos ["stroke"].astype ("float").mean (axis = 0)
dfAutos ["stroke"].replace (np.nan, prom, inplace = True)
prom = dfAutos ["horsepower"].astype ("float").mean (axis = 0)
dfAutos ["hoursepower"].replace (np.nan, prom, inplace = True)
prom = dfAutos ["peak"].astype ("float").mean (axis = 0)
dfAutos ["peak"].replace (np.nan, prom, inplace = True)
```

<h3 Style = "color: #8FBC8F"> 2.4 Imputación simple de valores categoricos: </h3>

```python
dfAutos ["doors"].value_counts ().idxmax ()
```

```python
dfAutos ["doors"].replace (np.nan, "four", inplace = True)
```

<h3 Style = "color: #8FBC8F"> 2.5 Eliminar filas con valores faltantes: </h3>

```python
dfAutos.dropna (subset = ["price"], axis = 0, inplace = True)
dfAutos.reset_index (drop = True, inplace = True)
```

```python
dfTemp = dfAutos.isnull()
for columna in dfTemp.columns.values.tolist ():
    print (columna)
    print (dfTemp[columna].value_counts())
    print ("")
```
